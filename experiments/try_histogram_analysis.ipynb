```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/no_padding/single/single-data-1731447983996344100.csv', delimiter=";")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='tries', bins=30, kde=True)
plt.title('Distribution of Number of Tries')
plt.xlabel('Number of Tries')
plt.ylabel('Frequency')
plt.grid(True)
```

```python
os.makedirs('../images', exist_ok=True)
```

```python
plt.savefig('../images/tries_histogram.png')
plt.close()
```

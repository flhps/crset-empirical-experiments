```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = '../out/benchStCas-varyRhatAndP-1729453982.38742.csv'
df = pd.read_csv(file_path, sep=';')
df = df.sort_values(['rhat', 'p'])
```

```python
fig = plt.figure(figsize=(12, 8))
plt.title('Mean Total Bitstring Length vs p for Different rhat Values')
    
for rhat in df['rhat'].unique():
    rhat_data = df[df['rhat'] == rhat]
    plt.plot(rhat_data['p'], rhat_data['bitsize'], 
                     label=f'rhat={rhat}')

plt.xlabel('p')
plt.ylabel('Mean Total Bitstring Length')
plt.legend()
plt.grid(True)
# plt.savefig('bitstring_length_vs_fpr.png')
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.title('Optimal p vs Different rhat Values')

data = []
for rhat in df['rhat'].unique():
    rhat_data = df[df['rhat'] == rhat].sort_values(['bitsize'])
    data.append([rhat, rhat_data.iloc[0]['p']])

df2 = pd.DataFrame(data,columns=['rhat', 'popt'])
plt.plot(df2['rhat'],df2['popt'])

plt.xlabel('rhat')
plt.ylabel('optimal p for min bitsize')
plt.legend()
plt.grid(True)
# plt.savefig('bitstring_length_vs_fpr.png')
plt.show()
```
